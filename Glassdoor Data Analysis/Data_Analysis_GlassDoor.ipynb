##a) Two Methods for handling missing data in Python

Handling missing data is a critical step in data preprocessing, affecting the quality and reliability of your analysis and modeling. Here are two common methods for handling missing data in Pandas DataFrames, along with examples for each:

### 1. Drop from Our Data Set
This method involves removing rows or columns that contain missing values. It's straightforward but can lead to loss of data, especially if many values are missing.

- **Example:**



```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
df = pd.DataFrame({'A': [1, 2, np.nan],
                   'B': [5, np.nan, np.nan],
                   'C': [1, 2, 3]})

# Drop rows where any cell in that row is NaN
df_dropna = df.dropna()

# Drop columns where any cell in that column is NaN
df.dropna(axis=1)
```

### 2. Replace the Missing Data with Another Value (Imputation)
This method involves filling in missing data with a specific value. Common fill values include the mean, median, or mode of the column.

- **Example:**

```python
# Fill missing values with the mean of the column
df['A'].fillna(value=df['A'].mean(), inplace=True)

# Fill missing values with a specific value
df.fillna(value=0)
```

## b) Methods for handling missing data in Python

In real-world data analysis, handling missing data is a fundamental step that significantly impacts the quality, reliability, and validity of the analysis. Missing data can arise due to various reasons, such as errors in data collection, non-responses in surveys, or disruptions in data transmission. The approach to handling missing data can vary depending on the nature of the data, the amount of missingness, and the analysis goals. Below, we discuss the significance, techniques, challenges, and implications of dealing with missing data in data analysis projects.

### Significance of Handling Missing Data

Handling missing data is crucial because:
- **Avoiding Bias:** Incorrectly handled missing data can introduce bias, leading to misleading results.
- **Ensuring Data Quality:** Proper handling improves the overall quality of the data set, making it more representative of the underlying population or phenomena.
- **Algorithm Compatibility:** Many machine learning algorithms require complete case analysis and cannot handle missing values directly.

### Techniques for Handling Missing Data

1. **Replace Missing Data with Another Value (Imputation):**
   - **Mean/Median/Mode Imputation:** Replace missing values with the mean, median, or mode of the column. Suitable for numerical data with a normal distribution or categorical data.
   - **Constant Value:** Replace missing data with a constant value that might represent an "unknown" category or a placeholder value.
   - **Example:** Filling missing survey responses with the most common response to maintain the dataset's size and structure.

2. **Fill in the Missing Data Using Existing Data:**
   - **Forward Fill/Backward Fill:** In time-series data, forward fill uses the previous data point to fill a missing value, while backward fill uses the next valid observation.
   - **Example:** In stock price data, using the last known price to fill a missing price for a day can maintain the continuity of data without introducing significant bias.

3. **Drop from Our Data Set:**
   - Dropping rows or columns with missing values. This method is straightforward but can lead to significant data loss.
   - **Example:** In a dataset with very few missing values concentrated in specific rows, removing these rows may have a minimal impact on the analysis.

4. **Interpolation:**
   - Estimating missing values using interpolation techniques, which can be linear or more complex (e.g., polynomial or spline).
   - **Example:** In environmental data, interpolating temperature readings between two time points can provide a reasonable estimate for missing data.

### Challenges and Implications

- **Data Loss:** Dropping too many rows or columns can significantly reduce the dataset size, potentially losing valuable information.
- **Bias Introduction:** Improper imputation can introduce bias, especially if the method does not reflect the data's true distribution or the nature of missingness.
- **Complexity in Decision Making:** Choosing the most appropriate method for handling missing data can be complex, requiring a deep understanding of the data and its context.

### Importance of Assessment and Documentation

- **Understanding Missingness:** Assessing the pattern and mechanism of missing data (MCAR, MAR, or NMAR) is critical for selecting an appropriate handling method.
- **Transparency and Reproducibility:** Documenting the chosen strategies for handling missing data, along with their rationale, ensures transparency and enables others to reproduce or build upon your analysis.

### Practical Examples

- **Healthcare Analysis:** Imputing missing values in patient data using median imputation for skewed distributions or mode imputation for categorical variables can prevent the loss of critical patient records.
- **Financial Time Series:** Using interpolation or forward fill to handle missing stock prices ensures continuity in data, crucial for trend analysis and forecasting models.

The strategy for handling missing data must be chosen carefully, considering the data's characteristics and the analysis objectives. Transparently documenting these decisions is essential for maintaining the integrity and reproducibility of the analysis.

## a) Creating Custom Dataset

```python
import random

random.seed(42)

# Define the possible values for each column
job_titles = ['Data Scientist', 'Data Engineer', 'Machine Learning Engineer', 'Data Analyst']
min_salaries = [31, 56, 66, 69, 80, 87, 90, 91, 101, 115, 110]

# Generate the lists for each column
job_title_list = random.choices(job_titles, k=20)  # Job Title
ratings = [round(random.uniform(0, 5), 1) for _ in range(20)]  # Rating
min_salary_list = random.choices(min_salaries, k=20)  # min_salary

# Assemble the dictionary with the generated lists
jobs_dict = {
    "Job Title": job_title_list,
    "Rating": ratings,
    "min_salary": min_salary_list
}

# The dictionary is now ready and contains 20 entries for each column as specified
custom_data = pd.DataFrame(jobs_dict)
custom_data
```

## b & c) Operations and Explainations

### **Left Join**

A Left join keeps all rows from the left dataset and matches rows from the right dataset. When there is no match, the result is NaN for the missing side in the output. This join keeps all rows from the original "Cleaned_DS_Jobs" dataset and adds matching rows from our newly created dataset based on the `Job Title` column. In cases where there's no match for a `Job Title` from the new dataset in the original dataset, the new columns (`Rating_y` and `min_salary_y`) will have `NaN` values.

For instance, the first row corresponding to the `Sr Data Scientist` job title in the original dataset doesn't find an exact match in the new dataset (since "Sr Data Scientist" wasn't a part of the new dataset's `Job Title` list), leading to `NaN` values in the `Rating_y` and `min_salary_y` columns.

### Left Join Explanation:
- **What does this do?** A Left join ensures that all entries from the left dataframe (in this case, "Cleaned_DS_Jobs") are retained in the merged result. If there are matching values in the right dataframe's join key column, those rows are added to the result. If no matches are found, the new columns from the right dataframe are filled with `NaN` for those rows.


```python
# Perform a Left join
left_join_df = df.merge(custom_data, on='Job Title', how='left')

# Display the first few rows of the result to verify the join
left_join_df.head()
```

### **Right Join:**
This join keeps all rows from our newly created dataset and adds matching rows from the original "Cleaned_DS_Jobs" dataset based on the Job Title column. In cases where there's no match for a Job Title from the original dataset in our new dataset, those rows from the original dataset are not included in the final result.

### **Right Join Explaination:**
- **How is it different from a Left join?** A Right join ensures that all entries from the right dataframe (our newly created dataset) are retained in the merged result, opposite to a Left join which retains all entries from the left dataframe (the original "Cleaned_DS_Jobs"). In our case, every `Job Title` from the new dataset finds matches in the original dataset, hence they are included in the merged result. If there were job titles in the new dataset not found in the original, those entries would still appear in the result, with missing data from the original dataset marked as `NaN`.

- **When is a Right Join appropriate?** A Right Join is appropriate when the integrity of the right dataframe needs to be maintained in the merged dataset. For example, if we specifically wanted to ensure that every new `Job Title` we generated, regardless of whether it has a direct match in the Glassdoor dataset, appears in our final analysis, we would use a Right join. This could be useful in scenarios where we are adding new job titles (perhaps from a different source or newer data) to an existing dataset and want to ensure all new titles are represented in the final dataset, even if there's no existing data to match them to.

```python
# Perform a right join
right_join_df = df.merge(custom_data, on='Job Title', how='right')

# Display the first few rows of the result to verify the join
right_join_df.head()
```

### **Outer join:**
It combines all rows from both datasets, regardless of whether there are matches between them. If there are no matches for a particular row in one dataset, the resulting dataset will contain `NaN` values for the missing side.

### **Outer Join Explanations:**
- **When is this typically used in data merging?** An Outer join is typically used when you want to retain all information from both datasets being merged, regardless of whether there are matches between them. This is useful for getting a complete picture of both datasets, including the differences and similarities.

- **Example situation for choosing an Outer Join over other types:** Suppose you have two datasets from different time periods or sources regarding job listings. One dataset contains job listings from the last year, and the other is a newly scraped dataset with current listings. An Outer join would allow you to merge these datasets to see all job listings, including those that do not overlap (i.e., jobs that were only listed in one of the two datasets). This can help in analyzing trends over time, identifying new roles that have emerged, and roles that no longer appear in newer datasets.

```python
# Perform a outer join
right_join_df = df.merge(custom_data, on='Job Title', how='outer')

# Display the first few rows of the result to verify the join
right_join_df.head()
```

### **Inner Join:**
This join combines rows from both datasets where there are matching values in the join column(s). Rows that do not find a match in the other dataset are excluded from the result.

### **Inner Join Explanation:**
- **Example scenario for an Inner Join:** An Inner Join is most suitable when you only want to analyze data that exists in both datasets. For example, if you're examining the overlap between two datasets of job listings to understand which jobs are consistently in demand across both datasets, an Inner Join would give you exactly those jobs listed in both, excluding any that are unique to one dataset or the other.

- **Potential challenges or drawbacks:**
  - **Data Loss:** One of the main challenges of using an Inner Join is the potential loss of data. Rows that do not match between the datasets are not included in the final merged dataset, which might exclude relevant data from the analysis.
  - **Assumption of Completeness:** It assumes that if data does not exist in both datasets, it is not relevant for the analysis. This might not always be true, especially if one of the datasets has missing information or if there are discrepancies in the data entry that prevent matches from being identified.

```python
# Perform a inner join
right_join_df = df.merge(custom_data, on='Job Title', how='inner')

# Display the first few rows of the result to verify the join
right_join_df.head()
```
